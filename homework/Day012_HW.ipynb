### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

# print (r.text)
soup = BeautifulSoup(r.text, "html5lib")

# 取出今天所有的發文
for d in soup.find(class_="part_list_2").find_all('h3'):
    print (d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
# 如果想要依照類別分類，怎麼儲存會比較好？
data = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    # print(d.find(class_="date").text, d.find(class_="tag").text, d.find_all('a')[-1].text)
    data.setdefault(d.find(class_="tag").text, [])
    data[d.find(class_="tag").text].append({'date' : d.find(class_="date").text,'title' : d.find_all('a')[-1].text})

for tag, value in data.items():
    for i in range(len(value)): 
        print ("tag : ", tag, "value : ", value[i])
```

### ③ 哪一個類別的文章最多

```python
# 哪一個類別的文章最多
top_count=0
for tag, value in data.items():
    if len(value) > top_count:
        top_count = len(value)
        
for tag, value in data.items():
    if len(value) == top_count:
        print("熱門類別 : ", tag, "文章數量 : ", top_count)
```
