## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 引入函式庫
import requests
import json
# 想要爬資料的⽬目標網址
# https://www.dcard.tw/_api/forums/pet/posts?popular=true
r = requests.get('https://www.dcard.tw/_api/forums/job/posts?popular=true')
response = r.text
# 利用 JSON 解析網頁內容
data = json.loads(response)

# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

print("Dcard API 回傳", len(data), "筆資料。")
rowNum = 0
for row in data:
    rowNum = rowNum + 1
    print("=====================================================================", rowNum)
    fieldCount = 0
    for key in row:
        fieldCount = fieldCount + 1
        print (fieldCount, key)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
rowNum = 0
for row in data:
    rowNum = rowNum + 1
    print(rowNum, row['title'], row['createdAt'], row['commentCount'], row['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

# 熱門文章
r = requests.get('https://www.dcard.tw/_api/forums/job/posts?popular=true')
response = r.text
data = json.loads(response)
TopCommentCount = 0
TopLikeCount = 0
for row in data:
    TopCommentCount = TopCommentCount + row['commentCount']
    TopLikeCount = TopLikeCount + row['likeCount']

print ("熱門文章 : 「平均留言人數」" + str(TopCommentCount / len(data)))
print ("熱門文章 : 「平均按讚人數」" + str(TopLikeCount / len(data)))

# 非熱門文章
r = requests.get('https://www.dcard.tw/_api/forums/job/posts?popular=false')
response = r.text
data = json.loads(response)
UnTopCommentCount = 0
UnTopLikeCount = 0
for row in data:
    UnTopCommentCount = UnTopCommentCount + row['commentCount']
    UnTopLikeCount = UnTopLikeCount + row['likeCount']

print ("非熱門文章 : 「平均留言人數」" + str(UnTopCommentCount / len(data)))
print ("非熱門文章 : 「平均按讚人數」" + str(UnTopLikeCount / len(data)))
```
