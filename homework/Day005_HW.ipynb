```python
# å¼•å…¥å‡½å¼åº«
import requests
import json
# æƒ³è¦çˆ¬è³‡æ–™çš„â½¬ç›®æ¨™ç¶²å€
# https://www.dcard.tw/_api/forums/pet/posts?popular=true
r = requests.get('https://www.dcard.tw/_api/forums/job/posts?popular=true')
response = r.text
# åˆ©ç”¨ JSON è§£æç¶²é å…§å®¹
data = json.loads(response)

# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

print("Dcard API å›å‚³", len(data), "ç­†è³‡æ–™ã€‚")
rowNum = 0
for row in data:
    rowNum = rowNum + 1
    print("=====================================================================", rowNum)
    fieldCount = 0
    for key in row:
        fieldCount = fieldCount + 1
        print (fieldCount, key)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
rowNum = 0
for row in data:
    rowNum = rowNum + 1
    print(rowNum, row['title'], row['createdAt'], row['commentCount'], row['likeCount'])
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

# ç†±é–€æ–‡ç« 
r = requests.get('https://www.dcard.tw/_api/forums/job/posts?popular=true')
response = r.text
data = json.loads(response)
TopCommentCount = 0
TopLikeCount = 0
for row in data:
    TopCommentCount = TopCommentCount + row['commentCount']
    TopLikeCount = TopLikeCount + row['likeCount']

print ("ç†±é–€æ–‡ç«  : ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€" + str(TopCommentCount / len(data)))
print ("ç†±é–€æ–‡ç«  : ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€" + str(TopLikeCount / len(data)))

# éç†±é–€æ–‡ç« 
r = requests.get('https://www.dcard.tw/_api/forums/job/posts?popular=false')
response = r.text
data = json.loads(response)
UnTopCommentCount = 0
UnTopLikeCount = 0
for row in data:
    UnTopCommentCount = UnTopCommentCount + row['commentCount']
    UnTopLikeCount = UnTopLikeCount + row['likeCount']

print ("éç†±é–€æ–‡ç«  : ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€" + str(UnTopCommentCount / len(data)))
print ("éç†±é–€æ–‡ç«  : ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€" + str(UnTopLikeCount / len(data)))
```
