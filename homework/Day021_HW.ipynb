## 1. 取出今天所有的新聞

```python
import requests
from bs4 import BeautifulSoup
import datetime
import re

sysTime = datetime.datetime.now()
dateString = sysTime.strftime("%Y") + "-" + str(int(sysTime.strftime("%m"))) + "-" + str(int(sysTime.strftime("%d")))
# https://www.ettoday.net/news/news-list-2020-2-5-0.htm
url = "https://www.ettoday.net/news/news-list-" + dateString + "-0.htm"
# 打開瀏覽器

from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get(url)

# 每個兩秒鐘自動往下滑

import time
for i in range(60):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
# 關閉瀏覽器
browser.quit();

# 取出今天所有的新聞
pattern = re.compile(sysTime.strftime("%Y/%m/%d"))
print ("# 取出今天所有的新聞")
for d in soup.find(class_="part_list_2").find_all('h3'):
    if (re.search(pattern, d.find(class_="date").text)):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
# 取出現在時間兩小時內的新聞
dateTime_B2H = sysTime + datetime.timedelta(hours=-1)
pattern1 = re.compile(sysTime.strftime("%Y/%m/%d %H"))
pattern2 = re.compile(dateTime_B2H.strftime("%Y/%m/%d %H"))
print ("# 取出現在時間兩小時內的新聞")
for d in soup.find(class_="part_list_2").find_all('h3'):
    if (re.search(pattern1, d.find(class_="date").text) or re.search(pattern2, d.find(class_="date").text)):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
